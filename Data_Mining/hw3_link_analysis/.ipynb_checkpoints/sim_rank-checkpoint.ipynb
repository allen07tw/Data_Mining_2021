# DATA process

```python
import sys
import numpy as np
import pandas as pd
import networkx as nx

graph_1_list = []
graph_2_list = []
graph_3_list = []
graph_4_list = []
graph_5_list = []
graph_6_list = []

with open('graph_1.txt','r') as f:
    for line in f:
        graph_1_list.append(list(line.strip('\n').split(',')))
with open('graph_2.txt','r') as f:
    for line in f:
        graph_2_list.append(list(line.strip('\n').split(',')))
with open('graph_3.txt','r') as f:
    for line in f:
        graph_3_list.append(list(line.strip('\n').split(',')))
with open('graph_4.txt','r') as f:
    for line in f:
        graph_4_list.append(list(line.strip('\n').split(',')))
with open('graph_5.txt','r') as f:
    for line in f:
        graph_5_list.append(list(line.strip('\n').split(',')))
with open('graph_6.txt','r') as f:
    for line in f:
        graph_6_list.append(list(line.strip('\n').split(',')))


data_txt = np.loadtxt('IBM_testing_dataset.txt')
IBM_testing_txtDF = pd.DataFrame(data_txt)
IBM_data = []

for IBM_item,customer in enumerate(IBM_testing_txtDF[0]):
    if len(IBM_data) < int(customer):
        IBM_data.append(list())
        IBM_data[int(customer)-1].append(str(IBM_testing_txtDF[2][IBM_item]))
    else:
        IBM_data[int(customer)-1].append(str(IBM_testing_txtDF[2][IBM_item]))
```

# graph_1_C=0.3

```python
import numpy as np
from itertools import combinations
import time

start_time = time.time()

want_threshold_error = 0.0001
want_list = graph_1_list
decay_factor_C = 0.3

def find_dim(wnat_list):
    all_number_list = []
    for x in want_list:
        for num in x:
            all_number_list.append(int(num))
    dim = max(all_number_list)
    return dim

dim = find_dim(want_list)
adj_matrix_hub = np.zeros((dim, dim))
sim_combinations = list(combinations([x for x in range(dim)],2))

last_pagerank = np.ones(dim)/dim
# 建hub_matrix
for idx, hub_point in enumerate(want_list):
    y_location = int(hub_point[0]) - 1
    x_location = int(hub_point[1]) - 1
    adj_matrix_hub[y_location][x_location] += 1
    
last_sim_rank = np.eye(dim)
    
adj_matrix_authority = adj_matrix_hub.transpose()
times_graph_2 = 0    

while True:    
    times_graph_2 += 1
    new_sim_rank = np.zeros((dim,dim)) + last_sim_rank
    
    

    for combination in sim_combinations: #i代表當前再算那個node
        idx_pos_1 = combination[0]
        idx_pos_2 = combination[1]
        
        parent_pos_1 = adj_matrix_authority[idx_pos_1]
        parent_pos_2 = adj_matrix_authority[idx_pos_2]
        
        sim_value = np.dot(parent_pos_1,parent_pos_2)
        
        parent_idx_pos_1 = []
        parent_idx_pos_2 = []
        #找parent inlink
        for idx in range(len(adj_matrix_authority[idx_pos_1])):
            if adj_matrix_authority[idx_pos_1][idx] != 0:
                parent_idx_pos_1.append(idx)
        for idx in range(len(adj_matrix_authority[idx_pos_2])):
            if adj_matrix_authority[idx_pos_2][idx] != 0:
                parent_idx_pos_2.append(idx)
#         print(combination)
#         print(parent_idx_pos_1)
#         print(parent_idx_pos_2)
        if (len(parent_idx_pos_1) == 0) or (len(parent_idx_pos_2) == 0):
            sum_parent_simrank = 0
            
        else:
            sum_parent_simrank = decay_factor_C / len(parent_idx_pos_1) / len(parent_idx_pos_2)
        
        
        sim_rank_decay = sum_parent_simrank * sim_value
        new_sim_rank[idx_pos_1][idx_pos_2] = sim_rank_decay
        
    
    new_sim_rank_value = np.sum(np.absolute(new_sim_rank))
    last_sim_rank_value = np.sum(np.absolute(last_sim_rank))
    new_sim_rank = new_sim_rank/new_sim_rank_value
    last_sim_rank_nomalize = last_sim_rank/last_sim_rank_value
    dif_sim_rank = np.sum(np.absolute(new_sim_rank - last_sim_rank_nomalize)) 
#     print(new_sim_rank_value)
#     print(dif_sim_rank)
    if dif_sim_rank < want_threshold_error:
        print(f"C = {decay_factor_C}")
        print(f'{times_graph_2}')
        print(dif_sim_rank)
        print(len(new_sim_rank))
        print(new_sim_rank)
        end_time = time.time()
        spend_time = end_time - start_time 
        print(f'time cost {spend_time}')
#         np.savetxt(f"graph_1_SimRank_C={decay_factor_C}.txt", new_sim_rank, delimiter=' ',fmt = '%10.5f')
        break
    
    last_sim_rank = new_sim_rank
```

```python
import numpy as np
from itertools import combinations
import time

start_time = time.time()

want_threshold_error = 0.0001
want_list = graph_5_list
decay_factor_C = 0.7

def find_dim(wnat_list):
    all_number_list = []
    for x in want_list:
        for num in x:
            all_number_list.append(int(num))
    dim = max(all_number_list)
    return dim

dim = find_dim(want_list)
adj_matrix_hub = np.zeros((dim, dim))
sim_combinations = list(combinations([x for x in range(dim)],2))

last_pagerank = np.ones(dim)/dim
# 建hub_matrix
for idx, hub_point in enumerate(want_list):
    y_location = int(hub_point[0]) - 1
    x_location = int(hub_point[1]) - 1
    adj_matrix_hub[y_location][x_location] += 1
    
last_sim_rank = np.eye(dim)
    
adj_matrix_authority = adj_matrix_hub.transpose()
times_graph_2 = 0    

while True:    
    times_graph_2 += 1
    new_sim_rank = np.zeros((dim,dim)) + last_sim_rank
    
    

    for combination in sim_combinations: #i代表當前再算那個node
        idx_pos_1 = combination[0]
        idx_pos_2 = combination[1]
        
        parent_pos_1 = adj_matrix_authority[idx_pos_1]
        parent_pos_2 = adj_matrix_authority[idx_pos_2]
        
        sim_value = np.dot(parent_pos_1,parent_pos_2)
        
        parent_idx_pos_1 = []
        parent_idx_pos_2 = []
        #找parent inlink
        for idx in range(len(adj_matrix_authority[idx_pos_1])):
            if adj_matrix_authority[idx_pos_1][idx] != 0:
                parent_idx_pos_1.append(idx)
        for idx in range(len(adj_matrix_authority[idx_pos_2])):
            if adj_matrix_authority[idx_pos_2][idx] != 0:
                parent_idx_pos_2.append(idx)
#         print(combination)
#         print(parent_idx_pos_1)
#         print(parent_idx_pos_2)
        if (len(parent_idx_pos_1) == 0) or (len(parent_idx_pos_2) == 0):
            sum_parent_simrank = 0
            
        else:
            sum_parent_simrank = decay_factor_C / len(parent_idx_pos_1) / len(parent_idx_pos_2)
        
        
        sim_rank_decay = sum_parent_simrank * sim_value
        new_sim_rank[idx_pos_1][idx_pos_2] = sim_rank_decay
        
    
    new_sim_rank_value = np.sum(np.absolute(new_sim_rank))
    last_sim_rank_value = np.sum(np.absolute(last_sim_rank))
    new_sim_rank = new_sim_rank/new_sim_rank_value
    last_sim_rank_nomalize = last_sim_rank/last_sim_rank_value
    dif_sim_rank = np.sum(np.absolute(new_sim_rank - last_sim_rank_nomalize)) 
#     print(new_sim_rank_value)
#     print(dif_sim_rank)
    if dif_sim_rank < want_threshold_error:
        print(f'{times_graph_2}')
        print(dif_sim_rank)
        print(len(new_sim_rank))
        print(new_sim_rank)
        end_time = time.time()
        spend_time = end_time - start_time 
        print(f'time cost {spend_time}')
        np.savetxt(f"graph_5_SimRank_C={decay_factor_C}.txt", new_sim_rank, delimiter=' ',fmt = '%10.5f')
        break
    
    last_sim_rank = new_sim_rank
```
