# plot groceries.csv distribution

```python
'''
Generates frequent item sets and association rules from given datasets using Apriori algorithm.
'''
# pylint: disable=invalid-name
import csv
import itertools
import time
import matplotlib.pyplot as plt
plt.rc('font', family='serif', size=10) 

# DataFile = open('Market_Basket_Optimisation.csv', 'r')
DataFile = open('groceries.csv', 'r')
DataCaptured = csv.reader(DataFile, delimiter=',')
data = list(DataCaptured)
print(len(data))
minsup = float(input("Support-Threshold: ")) #某物含的%數
f2 = "Apriori_Rules.txt"
f1 = "Apriori_frequent_pattern.txt"
minconf = float(input("Confidence-Threshold: "))
time_start = time.time()


def L1(data): #計算L1
    '''
    Find frequent 1-itemsets0
    '''
    #Get all 1-itemsets in the list items and their counts in the dictionary counts
    row_len= len(data)
    count = {}
    for items in data:
        for item in items:
            if item not in count:                   #計算個別總量
                count[(item)] = 1
            else:
                count[(item)] += 1
    count2 = {k: v for k, v in count.items()}
    plt.plot(range(len(count2)),[0.0045 *len(data)] * (len(count2)), 'r')
    plt.plot(range(len(count2)),[0.045 *len(data)] * (len(count2)), 'y')
    plt.scatter(range(len(count2)), list(count2.values()))
#     plt.bar(range(len(count2)), list(count2.values()), align='center', width=0.1)
    plt.xticks(range(len(count2)), list(sorted(count2.keys())))
    plt.xlabel('items_ groceries', fontsize = 16)
    plt.ylabel('times', fontsize = 16)
    plt.savefig('L_ groceries.jpg')
    plt.show()
    return count2


def apriori():
    '''
    The runner function
    '''    
    L = L1(data)
    flag = 1
    FreqItems = dict(L)
 
if __name__ == '__main__':
    apriori()
time_end = time.time()
print('Apriori_spent_time', time_end - time_start)
#     print(time_end - time_start)
```
