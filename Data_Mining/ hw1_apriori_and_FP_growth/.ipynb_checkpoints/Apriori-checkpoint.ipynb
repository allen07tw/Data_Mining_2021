# Apriori_test with groceries.csv¶

```python
'''
Generates frequent item sets and association rules from given datasets using Apriori algorithm.
'''
# pylint: disable=invalid-name
import csv
import itertools
import time



# DataFile = open('Market_Basket_Optimisation.csv', 'r')
DataFile = open('groceries.csv', 'r')
DataCaptured = csv.reader(DataFile, delimiter=',')
data = list(DataCaptured)
minsup = float(input("Support-Threshold: ")) #某物含的%數
f2 = "Apriori_Rules.txt"
f1 = "Apriori_frequent_pattern.txt"
minconf = float(input("Confidence-Threshold: "))
time_start = time.time()


def L1(data): #計算L1
    '''
    Find frequent 1-itemsets
    '''
    #Get all 1-itemsets in the list items and their counts in the dictionary counts
    row_len= len(data)
    count = {}
    for items in data:
        for item in items:
            if item not in count:                   #計算個別總量
                count[(item)] = 1
            else:
                count[(item)] += 1
    count2 = {k: v for k, v in count.items() if v >= minsup*row_len} #7501CSV行數，符合minsup我才要
    return count2


def generateCk(Lk_1, flag, data):
    '''
    Generate Ck by joining 2 Lk-1
    '''
    Ck = []
    # lk_1 = [a,b,c,e] 
    if flag == 1:
        flag = 0
        Ck += [(item1, item2) for item1 in Lk_1 for item2 in Lk_1 if item2 > item1]
    else:
        for item in Lk_1: #L2 = [('a', 'b'), ('a', 'c'), ('a', 'e'), ('b', 'c'), ('b', 'e'), ('c', 'e')]
            k = len(item)
        for item1 in Lk_1:
            for item2 in Lk_1:
                if (item1[:-1] == item2[:-1]) and (item1[-1] != item2[-1]): 
                    if item1[-1] > item2[-1]:
                        Ck.append(item2 + (item1[-1],)) #解決(a,b,b,c,)問題，指讓(a,b,c)存在 ，此外幣使用tuple才能產生('a','b')的key
                    else:                        
                        Ck.append(item1 + (item2[-1],)) #list是不行的
    L = generateLk(set(Ck), data)
    return L, flag


def generateLk(Ck, data):
    '''
    If item in Ck belongs to a transaction,
    it makes it into list Ct
    Then Ct is thresholded to form L
    '''
    row_len= len(data)
    count = {}
    for itemset in Ck:     #Ck = [(a,b),(a,c),(a,e), (b,c) ,(b,e),(c,e)]
        #print(itemset)
        for transaction in data: #data = list(DataCaptured)
            if all(e in transaction for e in itemset): # e = a , b itemset = (a,b)
                if itemset not in count: 
                    count[itemset] = 1
                else:
                    count[itemset] += 1
    count2 = {k: v for k, v in count.items() if v >= minsup*row_len}
    return count2


def rulegenerator(fitems): #fitmes = 最終dic
    '''
    Generates association rules from the frequent itemsets
    '''
    counter = 0
    for itemset in fitems.keys():
        if isinstance(itemset, str): #只有單個得我不要，我要tuple形式('a','b')
            continue
        length = len(itemset)#tuple裡面有幾項
        union_support = fitems[tuple(itemset)]#原本itemset為LIST形式，必須使用tuple(才可以變成key ，看兩個以上的品項的出現次數
        for i in range(1, length):
            lefts = map(list, itertools.combinations(itemset, i)) #['mineral water'], ['pancakes']
            for left in lefts:
                if len(left) == 1:
                    if ''.join(left) in fitems:
                        leftcount = fitems[''.join(left)]#單品項出現次數 
                        conf = union_support / leftcount
                if conf >= minconf:
                    fo = open(f2, "a+")
                    right = list(itemset[:])
                    for e in left:                          #去除被分配在左邊的品項
                        right.remove(e)     
#                     fo.write(str(left) + ' (' + str(leftcount) + ')' + ' -> ' + str(right) + ' (' + str(fitems[''.join(right)]) + ')' + ' [' + str(conf) + ']' + '\n')
                    print(str(left) + ' -> ' + str(right) + ' (' + str(conf) + ')')
                    counter += 1
                    fo.close()
    print(counter, "rules generated")


def apriori():
    '''
    The runner function
    '''    
    L = L1(data)
    flag = 1
    FreqItems = dict(L)
    while(len(L) != 0): #算到字典數目不符合，minsup
#         print(L,len(L))
        fo = open(f1, "a+")
        for k, v in L.items():
            fo.write(str(k) + ' >>> ' + str(v) + '\n\n')
        fo.close()
        L, flag = generateCk(L, flag, data)
        FreqItems.update(L) #update = 字典的append
    print("Frequent_Patterns")
    print('number_of_frequent_pattern',len(FreqItems))
    print(FreqItems.keys())
    print("Association_Rules")
    rulegenerator(FreqItems)

if __name__ == '__main__':
    apriori()
time_end = time.time()
print('Apriori_spent_time', time_end - time_start)
#     print(time_end - time_start)
```

```python
'''
Generates frequent item sets and association rules from given datasets using Apriori algorithm.
'''
# pylint: disable=invalid-name
import csv
import itertools
import time



# DataFile = open('Market_Basket_Optimisation.csv', 'r')
DataFile = open('groceries.csv', 'r')
DataCaptured = csv.reader(DataFile, delimiter=',')
data = list(DataCaptured)
minsup = float(input("Support-Threshold: ")) #某物含的%數
f2 = "Apriori_Rules.txt"
f1 = "Apriori_frequent_pattern.txt"
minconf = float(input("Confidence-Threshold: "))
time_start = time.time()


def L1(data): #計算L1
    '''
    Find frequent 1-itemsets
    '''
    #Get all 1-itemsets in the list items and their counts in the dictionary counts
    row_len= len(data)
    count = {}
    for items in data:
        for item in items:
            if item not in count:                   #計算個別總量
                count[(item)] = 1
            else:
                count[(item)] += 1
    count2 = {k: v for k, v in count.items() if v >= minsup*row_len} #7501CSV行數，符合minsup我才要
    return count2


def generateCk(Lk_1, flag, data):
    '''
    Generate Ck by joining 2 Lk-1
    '''
    Ck = []
    # lk_1 = [a,b,c,e] 
    if flag == 1:
        flag = 0
        Ck += [(item1, item2) for item1 in Lk_1 for item2 in Lk_1 if item2 > item1]
    else:
        for item in Lk_1: #L2 = [('a', 'b'), ('a', 'c'), ('a', 'e'), ('b', 'c'), ('b', 'e'), ('c', 'e')]
            k = len(item)
        for item1 in Lk_1:
            for item2 in Lk_1:
                if (item1[:-1] == item2[:-1]) and (item1[-1] != item2[-1]): 
                    if item1[-1] > item2[-1]:
                        Ck.append(item2 + (item1[-1],)) #解決(a,b,b,c,)問題，指讓(a,b,c)存在 ，此外幣使用tuple才能產生('a','b')的key
                    else:                        
                        Ck.append(item1 + (item2[-1],)) #list是不行的
    L = generateLk(set(Ck), data)
    return L, flag


def generateLk(Ck, data):
    '''
    If item in Ck belongs to a transaction,
    it makes it into list Ct
    Then Ct is thresholded to form L
    '''
    row_len= len(data)
    count = {}
    for itemset in Ck:     #Ck = [(a,b),(a,c),(a,e), (b,c) ,(b,e),(c,e)]
        #print(itemset)
        for transaction in data: #data = list(DataCaptured)
            if all(e in transaction for e in itemset): # e = a , b itemset = (a,b)
                if itemset not in count: 
                    count[itemset] = 1
                else:
                    count[itemset] += 1
    count2 = {k: v for k, v in count.items() if v >= minsup*row_len}
    return count2


def rulegenerator(fitems): #fitmes = 最終dic
    '''
    Generates association rules from the frequent itemsets
    '''
    counter = 0
    for itemset in fitems.keys():
        if isinstance(itemset, str): #只有單個得我不要，我要tuple形式('a','b')
            continue
        length = len(itemset)#tuple裡面有幾項
        union_support = fitems[tuple(itemset)]#原本itemset為LIST形式，必須使用tuple(才可以變成key ，看兩個以上的品項的出現次數
        for i in range(1, length):
            lefts = map(list, itertools.combinations(itemset, i)) #['mineral water'], ['pancakes']
            for left in lefts:
                if len(left) == 1:
                    if ''.join(left) in fitems:
                        leftcount = fitems[''.join(left)]#單品項出現次數 
                        conf = union_support / leftcount
                if conf >= minconf:
                    fo = open(f2, "a+")
                    right = list(itemset[:])
                    for e in left:                          #去除被分配在左邊的品項
                        right.remove(e)     
#                     fo.write(str(left) + ' (' + str(leftcount) + ')' + ' -> ' + str(right) + ' (' + str(fitems[''.join(right)]) + ')' + ' [' + str(conf) + ']' + '\n')
                    print(str(left) + ' -> ' + str(right) + ' (' + str(conf) + ')')
                    counter += 1
                    fo.close()
    print(counter, "rules generated")


def apriori():
    '''
    The runner function
    '''    
    L = L1(data)
    flag = 1
    FreqItems = dict(L)
    while(len(L) != 0): #算到字典數目不符合，minsup
#         print(L,len(L))
        fo = open(f1, "a+")
        for k, v in L.items():
            fo.write(str(k) + ' >>> ' + str(v) + '\n\n')
        fo.close()
        L, flag = generateCk(L, flag, data)
        FreqItems.update(L) #update = 字典的append
    print("Frequent_Patterns")
    print('number_of_frequent_pattern',len(FreqItems))
    print(FreqItems.keys())
    print("Association_Rules")
    rulegenerator(FreqItems)

if __name__ == '__main__':
    apriori()
time_end = time.time()
print('Apriori_spent_time', time_end - time_start)
#     print(time_end - time_start)
```

```python
'''
Generates frequent item sets and association rules from given datasets using Apriori algorithm.
'''
# pylint: disable=invalid-name
import csv
import itertools
import time
import matplotlib.pyplot as plt

DataFile = open('Market_Basket_Optimisation.csv', 'r')
# DataFile = open('groceries.csv', 'r')
DataCaptured = csv.reader(DataFile, delimiter=',')
data = list(DataCaptured)
print(len(data))
minsup = float(input("Support-Threshold: ")) #某物含的%數
f2 = "Apriori_Rules.txt"
f1 = "Apriori_frequent_pattern.txt"
minconf = float(input("Confidence-Threshold: "))
time_start = time.time()


def L1(data): #計算L1
    '''
    Find frequent 1-itemsets
    '''
    #Get all 1-itemsets in the list items and their counts in the dictionary counts
    row_len= len(data)
    count = {}
    for items in data:
        for item in items:
            if item not in count:                   #計算個別總量
                count[(item)] = 1
            else:
                count[(item)] += 1
    count2 = {k: v for k, v in count.items() if v >= minsup*row_len} #7501CSV行數，符合minsup我才要
    return count2


def generateCk(Lk_1, flag, data):
    '''
    Generate Ck by joining 2 Lk-1
    '''
    Ck = []
    # lk_1 = [a,b,c,e] 
    if flag == 1:
        flag = 0
        Ck += [(item1, item2) for item1 in Lk_1 for item2 in Lk_1 if item2 > item1]
    else:
        for item in Lk_1: #L2 = [('a', 'b'), ('a', 'c'), ('a', 'e'), ('b', 'c'), ('b', 'e'), ('c', 'e')]
            k = len(item)
        for item1 in Lk_1:
            for item2 in Lk_1:
                if (item1[:-1] == item2[:-1]) and (item1[-1] != item2[-1]): 
                    if item1[-1] > item2[-1]:
                        Ck.append(item2 + (item1[-1],)) #解決(a,b,b,c,)問題，指讓(a,b,c)存在 ，此外幣使用tuple才能產生('a','b')的key
                    else:                        
                        Ck.append(item1 + (item2[-1],)) #list是不行的
    L = generateLk(set(Ck), data)
    return L, flag


def generateLk(Ck, data):
    '''
    If item in Ck belongs to a transaction,
    it makes it into list Ct
    Then Ct is thresholded to form L
    '''
    row_len= len(data)
    count = {}
    for itemset in Ck:     #Ck = [(a,b),(a,c),(a,e), (b,c) ,(b,e),(c,e)]
        #print(itemset)
        for transaction in data: #data = list(DataCaptured)
            if all(e in transaction for e in itemset): # e = a , b itemset = (a,b)
                if itemset not in count: 
                    count[itemset] = 1
                else:
                    count[itemset] += 1
#     if ('ham', 'whole wheat rice') in count:
#         print("('ham', 'whole wheat rice')",count[('ham', 'whole wheat rice')])
    count2 = {k: v for k, v in count.items() if v >= minsup*row_len}
    return count2


def rulegenerator(fitems): #fitmes = 最終dic
    '''
    Generates association rules from the frequent itemsets
    '''
    counter = 0
    for itemset in fitems.keys():
        if isinstance(itemset, str): #只有單個得我不要，我要tuple形式('a','b')
            continue
        length = len(itemset)#tuple裡面有幾項
        union_support = fitems[tuple(itemset)]#原本itemset為LIST形式，必須使用tuple(才可以變成key ，看兩個以上的品項的出現次數
        for i in range(1, length):
            lefts = map(list, itertools.combinations(itemset, i)) #['mineral water'], ['pancakes']
            for left in lefts:
                if len(left) == 1:
                    if ''.join(left) in fitems:
                        leftcount = fitems[''.join(left)]#單品項出現次數 
                        conf = union_support / leftcount
                if conf >= minconf:
                    fo = open(f2, "a+")
                    right = list(itemset[:])
                    for e in left:                          #去除被分配在左邊的品項
                        right.remove(e)     
#                     fo.write(str(left) + ' (' + str(leftcount) + ')' + ' -> ' + str(right) + ' (' + str(fitems[''.join(right)]) + ')' + ' [' + str(conf) + ']' + '\n')
                    print(str(left) + ' -> ' + str(right) + ' (' + str(conf) + ')')
                    counter += 1
                    fo.close()
    print(counter, "rules generated")


def apriori():
    '''
    The runner function
    '''    
    L = L1(data)
    flag = 1
    FreqItems = dict(L)
    while(len(L) != 0): #算到字典數目不符合，minsup
#         print(L,len(L))
        fo = open(f1, "a+")
        for k, v in L.items():
            fo.write(str(k) + ' >>> ' + str(v) + '\n\n')
        fo.close()
        L, flag = generateCk(L, flag, data)
        FreqItems.update(L) #update = 字典的append
    print("Frequent_Patterns")
    print('number of frquent_pattern',len(FreqItems))
    print(FreqItems.keys())
    print("Association_Rules")
    rulegenerator(FreqItems)
#     plt.bar(range(len(FreqItems)), list(FreqItems.values()), align='center')
#     plt.xticks(range(len(FreqItems)), list(FreqItems.keys()))
#     plt.show()
if __name__ == '__main__':
    apriori()
time_end = time.time()
print('Apriori_spent_time', time_end - time_start)
#     print(time_end - time_start)
```

# Apriori  big_minsup and big_minconf with Market_Basket_Optimisation.csv

```python
'''
Generates frequent item sets and association rules from given datasets using Apriori algorithm.
'''
# pylint: disable=invalid-name
import csv
import itertools
import time



DataFile = open('Market_Basket_Optimisation.csv', 'r')
# DataFile = open('', 'r')
DataCaptured = csv.reader(DataFile, delimiter=',')
data = list(DataCaptured)
minsup = float(input("Support-Threshold: ")) #某物含的%數
f2 = "Apriori_Rules.txt"
f1 = "Apriori_frequent_pattern.txt"
minconf = float(input("Confidence-Threshold: "))
time_start = time.time()


def L1(data): #計算L1
    '''
    Find frequent 1-itemsets
    '''
    #Get all 1-itemsets in the list items and their counts in the dictionary counts
    row_len= len(data)
    count = {}
    for items in data:
        for item in items:
            if item not in count:                   #計算個別總量
                count[(item)] = 1
            else:
                count[(item)] += 1
    count2 = {k: v for k, v in count.items() if v >= minsup*row_len} #7501CSV行數，符合minsup我才要
    return count2


def generateCk(Lk_1, flag, data):
    '''
    Generate Ck by joining 2 Lk-1
    '''
    Ck = []
    # lk_1 = [a,b,c,e] 
    if flag == 1:
        flag = 0
        Ck += [(item1, item2) for item1 in Lk_1 for item2 in Lk_1 if item2 > item1]
    else:
        for item in Lk_1: #L2 = [('a', 'b'), ('a', 'c'), ('a', 'e'), ('b', 'c'), ('b', 'e'), ('c', 'e')]
            k = len(item)
        for item1 in Lk_1:
            for item2 in Lk_1:
                if (item1[:-1] == item2[:-1]) and (item1[-1] != item2[-1]): 
                    if item1[-1] > item2[-1]:
                        Ck.append(item2 + (item1[-1],)) #解決(a,b,b,c,)問題，指讓(a,b,c)存在 ，此外幣使用tuple才能產生('a','b')的key
                    else:                        
                        Ck.append(item1 + (item2[-1],)) #list是不行的
    L = generateLk(set(Ck), data)
    return L, flag


def generateLk(Ck, data):
    '''
    If item in Ck belongs to a transaction,
    it makes it into list Ct
    Then Ct is thresholded to form L
    '''
    row_len= len(data)
    count = {}
    for itemset in Ck:     #Ck = [(a,b),(a,c),(a,e), (b,c) ,(b,e),(c,e)]
        #print(itemset)
        for transaction in data: #data = list(DataCaptured)
            if all(e in transaction for e in itemset): # e = a , b itemset = (a,b)
                if itemset not in count: 
                    count[itemset] = 1
                else:
                    count[itemset] += 1
    count2 = {k: v for k, v in count.items() if v >= minsup*row_len}
    return count2


def rulegenerator(fitems): #fitmes = 最終dic
    '''
    Generates association rules from the frequent itemsets
    '''
    counter = 0
    for itemset in fitems.keys():
        if isinstance(itemset, str): #只有單個得我不要，我要tuple形式('a','b')
            continue
        length = len(itemset)#tuple裡面有幾項
        union_support = fitems[tuple(itemset)]#原本itemset為LIST形式，必須使用tuple(才可以變成key ，看兩個以上的品項的出現次數
        for i in range(1, length):
            lefts = map(list, itertools.combinations(itemset, i)) #['mineral water'], ['pancakes']
            for left in lefts:
                if len(left) == 1:
                    if ''.join(left) in fitems:
                        leftcount = fitems[''.join(left)]#單品項出現次數 
                        conf = union_support / leftcount
                if conf >= minconf:
                    fo = open(f2, "a+")
                    right = list(itemset[:])
                    for e in left:                          #去除被分配在左邊的品項
                        right.remove(e)     
#                     fo.write(str(left) + ' (' + str(leftcount) + ')' + ' -> ' + str(right) + ' (' + str(fitems[''.join(right)]) + ')' + ' [' + str(conf) + ']' + '\n')
                    print(str(left) + ' -> ' + str(right) + ' (' + str(conf) + ')')
                    counter += 1
                    fo.close()
    print(counter, "rules generated")


def apriori():
    '''
    The runner function
    '''    
    L = L1(data)
    flag = 1
    FreqItems = dict(L)
    while(len(L) != 0): #算到字典數目不符合，minsup
#         print(L,len(L))
        fo = open(f1, "a+")
        for k, v in L.items():
            fo.write(str(k) + ' >>> ' + str(v) + '\n\n')
        fo.close()
        L, flag = generateCk(L, flag, data)
        FreqItems.update(L) #update = 字典的append
    print("Frequent_Patterns")
    print('number_of_frequent_pattern',len(FreqItems))
    print(FreqItems.keys())
    print("Association_Rules")
    rulegenerator(FreqItems)

if __name__ == '__main__':
    apriori()
time_end = time.time()
print('Apriori_spent_time', time_end - time_start)
#     print(time_end - time_start)
```
